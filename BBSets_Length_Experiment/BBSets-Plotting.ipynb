```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import ConnectionPatch

# ---------------------------
# USER-ADJUSTABLE PARAMETERS
# ---------------------------

# Define the rectangle (zoom) region on the left (ax1) that corresponds to the zoomed-in view on the right (ax2).
zoom_region = {
    'x_min': -0.001,
    'x_max': 0.05,
    'y_min': 3.1,
    'y_max': 3.6,
}

# Styling for the rectangle indicating the zoom region.
rectangle_params = {
    'linewidth': 1,
    'edgecolor': 'red',
    'facecolor': 'none',
    'linestyle': (0, (5, 5)),  # dash pattern: 5 on, 5 off
    'alpha': 0.5,
}

# Styling for the dashed lines that connect the zoom rectangle in ax1 to ax2.
line_params = {
    'color': 'red',
    'linewidth': 1,
    'linestyle': (0, (5, 5)),  # dash pattern: 5 on, 5 off
    'alpha': 0.7,
}

# ---------------------------
# MAIN PLOTTING CODE
# ---------------------------

# Ensure Î·s is defined. Replace the following line with your actual Î·s values.
Î·s = np.linspace(0, 0.32, 500)[1:-1]

# Define the runs and corresponding labels
name_of_runs = ['fixed_220_32', 'fixed_330_32', 'fixed_550_32']
x_star_labels = ['(2,2,0)', '(3,3,0)', '(5,5,0)']

# Set the plotting style once
plt.style.use('plot_style-coolwarm.txt')

# Create a figure with two subplots (panels)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=False)

# Loop through each run and plot on both subplots
for k, name_of_run in enumerate(name_of_runs):
    # Load the data
    lengths = np.load(f'lengths_final{name_of_run}.npy')
    
    # Compute mean and standard deviation
    mean_vector = np.mean(lengths, axis=0)

    std_dev = np.std(lengths, axis=0)
    
    # Define the label for the legend
    latex_part = r'$\mathbf{x^*}$'
    label = f'{latex_part} = {x_star_labels[k]}'
    
    # Plot on the right subplot (zoomed-in) with markers
    ax2.plot(
        [0] + list(Î·s),
        mean_vector,
        '-',
        linewidth = 2,
        #markersize=5,
        #mec='black',
        label=label
    )
    
    # Plot on the left subplot (full view) without markers
    ax1.plot(
        [0] + list(Î·s),
        mean_vector,
        '-',
        linewidth=2,
        label=label
    )

# Add a common title
fig.suptitle('Confidence Interval Lengths in the Berger-Boos Setting', fontsize=16)

# -----------------------------------------------
# Apply Zoom Settings (Rectangle + Connection)
# -----------------------------------------------

# Unpack zoom boundaries for convenience
x_zoom_min = zoom_region['x_min']
x_zoom_max = zoom_region['x_max']
y_zoom_min = zoom_region['y_min']
y_zoom_max = zoom_region['y_max']

# 1) Rectangle in ax1 indicating the zoom region
zoom_rect = Rectangle(
    (x_zoom_min, y_zoom_min),
    x_zoom_max - x_zoom_min,
    y_zoom_max - y_zoom_min,
    **rectangle_params  # Use the styling dictionary
)
ax1.add_patch(zoom_rect)

# 2) Connect the rectangle in ax1 to ax2 with dashed lines
#    We'll connect the top-left and bottom-left corners.
con_top = ConnectionPatch(
    xyA=(x_zoom_min, y_zoom_max),  # Point in ax1 data coordinates
    xyB=(0, 1),                    # Corresponding point in ax2 axes coordinates
    coordsA=ax1.transData,
    coordsB=ax2.transAxes,
    **line_params                  # Use the line styling dictionary
)
fig.add_artist(con_top)

con_bottom = ConnectionPatch(
    xyA=(x_zoom_min, y_zoom_min),  # Point in ax1 data coordinates
    xyB=(0, 0),                    # Corresponding point in ax2 axes coordinates
    coordsA=ax1.transData,
    coordsB=ax2.transAxes,
    **line_params
)
fig.add_artist(con_bottom)

# -----------------------------------------------
# Customize Axes
# -----------------------------------------------

# Right subplot (ax2) - zoomed-in region
ax2.set_xlim(x_zoom_min, x_zoom_max)
ax2.set_ylim(y_zoom_min, y_zoom_max)
ax2.set_xlabel(r'$\eta$')
ax2.set_ylabel('Average Interval length')
ax2.legend(loc='lower right')
ax2.grid(True, linestyle='--', alpha=0.5)

# Left subplot (ax1) - full range
ax1.set_xlim(-0.005, 0.32)
ax1.set_ylim(2.9, 4.6)
ax1.set_xlabel(r'$\eta$')
ax1.set_ylabel('Average Interval length')
ax1.legend(loc='lower right')
ax1.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to prevent overlap (leave space for title)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Save the combined figure (optional)
plt.savefig("BB_combined_figure.png", dpi=300)

# Display the figure
plt.show()
```
